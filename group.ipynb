```python
import pandas as pd
```

```python
path='Diabetes Dataset.csv'# 替换策划给你自己的地址
data=pd.read_csv(path)
type(data)
```

```python
data.head()
```

```python
print(data.shape)
data.describe()
```

可以看到读取的时候所有的特征除了前两个都存在999

```python
#缺失值和999的替代
null=data.isnull()
null
```

```python
null.sum()
```

```python
data=data.replace(' ',-1)
data.describe()
```

所有缺失值都已经替换为0,我们只需要替换999

```python
data=data.replace(999,0)
```

```python
import numpy as np
x=data.columns
y=data.describe().columns
print(np.array(x).shape)
print(np.array(y).shape)     #可以看到函数自动过滤掉了一部分特征，原因？
lost_column=set(x)-set(y)
print(lost_column)      #为什么这些列被过滤掉了
data['MeanPhysicianDistress']   #可以看到读取的数据类型是obiect
#数据格式也是混乱的，需要调整数据格式。
```

```python
data['CodeCentre']
```

```python
data=data[lost_column].astype(float)
```

```python
data['MeanPhysicianDistress'] 
###数据格式对了但是describe任然过滤掉了
```

```python
data.describe()
```

上面是一个直观的相关性色图，下面对部分特征做一个分布情况调研

```python
data.head()
import plotly.graph_objects as go
import numpy as np
name=data.columns
x_data = list(name)
fig = go.Figure()
for i in range(0,106):
    fig.add_trace(go.Histogram(x=data.iloc[:,i],
                               name=x_data[i])
                  )


fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.75)
fig.show()
```
